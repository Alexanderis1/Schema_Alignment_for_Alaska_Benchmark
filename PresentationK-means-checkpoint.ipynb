# Searching the best K for our Dataset based on Silhouette value

<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/3d80ab22fb291b347b2d9dc3cc7cd614f6b15479">

```python
silhouette_scores = {}
from sklearn import metrics
test = range(2,1000,10)

for k in test:
    model_kmeans_k = KMeans( n_clusters = k )
    model_kmeans_k.fit(df1.drop("cluster", axis = 1))
    labels_k = model_kmeans_k.labels_
    score_k = metrics.silhouette_score(df1.drop("cluster", axis=1), labels_k)
    silhouette_scores[k] = score_k
    print("Tested kMeans with k = %d\tSS: %5.4f" % (k, score_k))
```

```python
plt.figure(figsize = (40,10))
plt.plot(list(silhouette_scores.values()))
plt.xticks(range(0,100,1), silhouette_scores.keys())
plt.title("Silhouette Metric")
plt.xlabel("k")
plt.ylabel("Silhouette")
plt.axvline(12, color = "r")
plt.show()
```

```python
silhouette_scores = {}
from sklearn import metrics
test = range(120,142,1)

for k in test:
    model_kmeans_k = KMeans( n_clusters = k )
    model_kmeans_k.fit(df1.drop("cluster", axis = 1))
    labels_k = model_kmeans_k.labels_
    score_k = metrics.silhouette_score(df1.drop("cluster", axis=1), labels_k)
    silhouette_scores[k] = score_k
    print("Tested kMeans with k = %d\tSS: %5.4f" % (k, score_k))
```

```python
plt.figure(figsize = (40,10))
plt.plot(list(silhouette_scores.values()))
plt.xticks(range(0,22,1), silhouette_scores.keys())
plt.title("Silhouette Metric")
plt.xlabel("k")
plt.ylabel("Silhouette")
plt.axvline(7, color = "r")
plt.show()
```
